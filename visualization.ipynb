# 2025 Stats Visualizations
Focusing on 2025 first since it is more relevent. Looking at general teams and their statistics.

```python
def plot_per_set_stat(df, stat_name='Aces', team_col='Team', value_col='Per Set'):
    """
    Plots a bar chart showing 'Per Set' values (e.g., Aces, Blocks) for each volleyball team.

    """
    title = f'{stat_name} per Set by Team'
    ylabel = f'{stat_name} per Set'

    df_sorted = df.sort_values(by=value_col, ascending=False)

    plt.figure(figsize=(12, 6))
    plt.bar(df_sorted[team_col], df_sorted[value_col])
    plt.xlabel('Team')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

def plot_stacked_per_set_stats(df, team_col='Team', stat_cols=None, title='Per Set Stats by Team', color_map=None):
    """
    Plots a stacked bar chart for multiple 'Per Set' stats per team,
    preserving the original DataFrame row order.
    """
    if stat_cols is None:
        stat_cols = [col for col in df.columns if col != team_col]

    if color_map:
        colors = [color_map.get(stat, None) for stat in stat_cols]
    else:
        colors = None

    teams = df[team_col].tolist()
    x = np.arange(len(teams))

    plt.figure(figsize=(14, 6))
    bottom = np.zeros(len(df))

    for i, stat in enumerate(stat_cols):
        plt.bar(x, df[stat], bottom=bottom, label=stat, color=colors[i] if colors else None, edgecolor='black')
        bottom += df[stat].fillna(0).values  # safely handles NaNs



    plt.xticks(x, teams, rotation=45, ha='right')
    plt.title(title)
    plt.xlabel("Team")
    plt.ylabel("Per Set Value")
    plt.legend(title='Stat Type', loc='upper left', bbox_to_anchor=(1.05, 1))
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

def merge_per_set_dataframes(stat_dfs: dict):
    """
    Merges multiple 'Per Set' DataFrames on 'Team', using only the relevant columns.

    Parameters:
    - stat_dfs (dict): Keys are stat names (e.g., 'Aces'), values are DataFrames with 'Team' and 'Per Set' columns.
    
    Returns:
    - pd.DataFrame: Merged DataFrame with 'Team' and one column per stat.
    """
    merged_df = None

    for stat_name, df in stat_dfs.items():
        # Select only relevant columns, rename 'Per Set' to stat name
        df_clean = df[['Team', 'Per Set']].copy()
        df_clean = df_clean.rename(columns={'Per Set': stat_name})

        if merged_df is None:
            merged_df = df_clean
        else:
            merged_df = pd.merge(merged_df, df_clean, on="Team", how="outer")

    return merged_df

def merge_stats_with_win_pct(merged_stats_df, records_df, team_col='Team'):
    """
    Merges the per-set stat dataframe with the win/loss records dataframe on 'Team',
    and sorts by winning percentage (Pct.).
    """
    records_clean = records_df[[team_col, 'Pct.']].copy()

    full_df = pd.merge(merged_stats_df, records_clean, on=team_col, how='left')

    full_df = full_df.sort_values(by='Pct.', ascending=False)


    return full_df

```

```python
plot_per_set_stat(acesTeam2025, stat_name='Aces')
plot_per_set_stat(assistTeam2025, stat_name='Assists')
plot_per_set_stat(attacksTeam2025, stat_name='Attacks')
plot_per_set_stat(blocksTeam2025, stat_name='Blocks')
plot_per_set_stat(digsTeam2025, stat_name="Digs")
plot_per_set_stat(killsTeam2025, stat_name="Kills")
```

```python
color_map = {
    'Aces': '#1f77b4',
    'Blocks': '#ff7f0e',
    'Kills': '#2ca02c',
    'Assists': '#d62728',
    'Attacks': '#9467bd',
    'Digs': '#8c564b'
}

stat_dfs = {'Aces': acesTeam2025,'Blocks': blocksTeam2025,'Digs': digsTeam2025,'Kills': killsTeam2025,'Assists': assistTeam2025,'Attacks': attacksTeam2025}
merged_stats = merge_per_set_dataframes(stat_dfs)
# plot_stacked_per_set_stats(merged_stats, stat_cols=['Aces', 'Blocks', 'Kills', 'Assists', 'Attacks'])

sorted_stacked_per_set = merge_stats_with_win_pct(merged_stats, matchTeamPercentage2025)
stat_cols = list(stat_dfs.keys())
plot_stacked_per_set_stats(sorted_stacked_per_set, stat_cols=stat_cols, title="Sorted by Winning %", color_map=color_map)
```
