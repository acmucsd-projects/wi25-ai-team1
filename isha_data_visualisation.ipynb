```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
base_path = "allStatCSVS"

# Load team-level stats
team_stats_files = [
    "Match_W-L_Pctg._Team_2025.csv",
    "Aces_Per_Set_Team_2025.csv",
    "Assists_Per_Set_Team_2025.csv",
    "Blocks_Per_Set_Team_2025.csv",
    "Digs_Per_Set_Team_2025.csv"
]

# Load and merge team datasets
team_df = None
for file in team_stats_files:
    df = pd.read_csv(os.path.join(base_path, file))
    if team_df is None:
        team_df = df
    else:
        team_df = pd.merge(team_df, df, on="Team", how="inner")
```

```python
team_df.head()
```

```python
team_df["Wins"] = team_df["W"] / (team_df["L"] + team_df["W"])
```

```python
team_df["Lose"] = team_df["L"] / (team_df["L"] + team_df["W"])
```

```python
team_df.plot(x='Aces', y='Lose')
plt.show()
```

```python
team_df.plot(x='Digs', y='Ratio')
plt.show()
```

```python
team_df.plot(x='Digs', y='Wins')
plt.show()
```

```python
team_df.plot(x='Digs', y='Lose')
plt.show()
```

```python
corr_matrix = team_df.corr()

corr_matrix[['Wins', 'Lose']].sort_values(by='Wins', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix[['Wins', 'Lose']].sort_values('Wins', ascending=False), 
            annot=True, cmap='coolwarm', center=0)
plt.title('Correlation with Wins and Losses')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.barplot(x='Wins', y='Aces', data=team_df, ax=axes[0])
axes[0].set_title('Mutual Information with Wins')

sns.barplot(x='Lose', y='Aces', data=team_df.sort_values('Lose', ascending=False), ax=axes[1])
axes[1].set_title('Mutual Information with Losses')
plt.tight_layout()
plt.show()
```
