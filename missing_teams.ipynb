### Finding unique teams

```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('2024MensNcaaGames.csv')
```

```python
team_1 = set(df['Team 1'])
print(team_1)
```

```python
team_2 = set(df['Team 2'])
print(team_2)
```

```python
all_teams = team_1.union(team_2)
print(all_teams)
```

```python
import pandas as pd
import glob
import os

# Set the path where your CSV files are stored
folder_path = './allStatCSVS'

# Use glob to find all CSV files ending with 2024.csv
csv_files = glob.glob(os.path.join(folder_path, '*Team_2024.csv'))

# Iterate through each matching file
count = 0
for file in csv_files:
    if (count == 0):
        count +=1
        continue
    df = pd.read_csv(file)
    print(f"Processing file: {file}")
    # Do something with df here
    print(df.shape)
    for teamName in all_teams:
        q1 = df["Team"].apply(lambda x : x in teamName)
        q2 = df["Team"].str.contains(teamName)
        filteredDF = df[q1 | q2]
        display(filteredDF)
    break
```
