```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

### Read CSV files

```python
df_wins = pd.read_csv('allStatCSVS/Match_W-L_Pctg._Team_2025.csv')
df_aces_2025 = pd.read_csv('allStatCSVS/Aces_Per_Set_Team_2025.csv')
df_hits_2025 = pd.read_csv('allStatCSVS/Hitting_Percentage_Team_2025.csv')
df_kills_2025 = pd.read_csv('allStatCSVS/Kills_Per_Set_Team_2025.csv')
```

```python
# Group by team and sum aces
team_aces = df_aces_2025.groupby("Team", as_index=False)["Aces"].sum()
team_aces
```

```python
df_wins_aces = pd.merge(df_wins, df_aces_2025, on='Team')
```

```python
df_wins_aces
```

```python
sns.lineplot(data=df_wins_aces, x="Aces", y="Pct.")
plt.title("Team Aces vs Wins")
plt.xlabel("Total Team Aces")
plt.ylabel("Total Team Wins Percentage")
plt.show()
```
